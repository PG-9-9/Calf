```python
import shutil
import os
```

```python
def copy_files_by_date(source_dir, target_dir, dates):
    # Ensure the target directory exists
    os.makedirs(target_dir, exist_ok=True)

    # Iterate over all files in the source directory
    for filename in os.listdir(source_dir):
        # Check if the file name contains any of the specified dates
        if any(date in filename for date in dates):
            source_file = os.path.join(source_dir, filename)
            target_file = os.path.join(target_dir, filename)

            # Copy the file
            shutil.copy2(source_file, target_file)
            print(f"Copied {filename} to {target_dir}")
```

```python
source_directory = 'Calf_Detection/Audio/Audio_Work/wav_data'
target_directory = 'Calf_Detection/Audio/Audio_Work_AE/normal_calf_subset'
dates_to_copy = ['2023-10-08']

copy_files_by_date(source_directory, target_directory, dates_to_copy)
```

```python
import os
import shutil

def copy_files_except_dates(source_dir, output_dir, dates_not_to_copy):
    """
    Copies files from source directory to output directory, excluding files with specified dates in their names.

    :param source_dir: Path to the source directory containing the files.
    :param output_dir: Path to the output directory where files will be copied.
    :param dates_not_to_copy: List of dates in 'YYYY-MM-DD' format to be excluded.
    """
    # Ensure output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(source_dir):
        # Check if the file's date is not in the dates_not_to_copy list
        if not any(date in filename for date in dates_not_to_copy):
            # Construct full file path
            source_file = os.path.join(source_dir, filename)
            destination_file = os.path.join(output_dir, filename)

            # Copy the file
            shutil.copy2(source_file, destination_file)
            print(f"Copied: {filename}")

    print("File copying complete.")

```

```python
source_dir='Calf_Detection/Audio/Audio_Work/wav_data'
destination_dir='Calf_Detection/Audio/Audio_Work_AE/normal_calf_set'
dates_not_to_copy=['2023-10-08','2023-10-09','2023-10-11','2023-10-12','2023-10-17','2023-10-25','2023-10-31','2023-11-04']
copy_files_except_dates(source_dir, destination_dir, dates_not_to_copy)
```
