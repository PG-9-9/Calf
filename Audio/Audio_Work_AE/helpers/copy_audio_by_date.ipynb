```python
import os
import shutil
from datetime import datetime

def copy_files_for_specific_dates(source_dir, target_dir, dates_list):
    # Ensure the target directory exists
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    
    # Convert the dates in the list to datetime objects for comparison
    dates_objs = [datetime.strptime(date, '%Y-%m-%d') for date in dates_list]
    
    # List all files in the source directory
    for file in os.listdir(source_dir):
        # Check if the file matches the expected format
        if file.startswith('output_') and file.endswith('.wav'):
            # Extract the date from the file name
            date_str = file.split('_')[1]
            file_date_obj = datetime.strptime(date_str, '%Y-%m-%d')
            
            # Check if the file's date is within the specified dates list
            if file_date_obj in dates_objs:
                # Copy the file to the target directory
                shutil.copy(os.path.join(source_dir, file), target_dir)
                print(f"Copied: {file}")
```

```python

source_directory = '/home/woody/iwso/iwso122h/Calf_Detection/Audio/Audio_Work/wav_data'
target_directory = '/home/woody/iwso/iwso122h/Calf_Detection/Audio/Audio_Work_AE/test_set'
dates_list = ['2023-10-12', '2023-10-25','2023-10-31']

copy_files_for_specific_dates(source_directory, target_directory, dates_list)
```
